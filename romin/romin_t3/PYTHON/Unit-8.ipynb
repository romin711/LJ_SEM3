```python
def div(a,b):
    return a/b
try:
    print(div(5,0))
except Exception as e:
    print(e)
#  except:
#      print("exception")
else:
    print("No exception")
finally:
    print("Always run")
```

```python
def div(a,b):
    return a/b
try:
    print(div(5,2))
    a=[1,2,3]
    print(a[4])
except Exception as e:
    print(e)
else:
    print("No exception")
finally:
    print("Always run")
```

```python
def div(a,b):
    return a/b
try:
    print(div(5,2))
    a=[1,2,3]
    print(a[4])
except ZeroDivisionError:
    print("0 div")
except IndexError:
    print("index")
#     except (ZeroDivisionError,IndexError)
else:
    print("No exception")
finally:
    print("Always run")
```

# Custom Exception

```python
def add(a,b):
    if type(a)==str:
        raise TypeError("A must be int")
    if type(b)==str:
        raise TypeError("B must be int")
    return (a+b)
try:
    print(add("asd","rty"))
except Exception as e:
    print(e)
```
