```python
import matplotlib.pyplot as plt 
import numpy as np

# x = np.array([1,2,3,4,5])
# y = np.array([8,3,5,2,9])
# z = np.array([1,3,5,7,9])


x = np.arange(1,11)
y = np.arange(5,15)
z = np.arange(21,41,2)

plt.title('1st line graph',fontsize=20,color='green',fontweight='bold',fontfamily='serif')
plt.xlabel('x axis--->',fontsize=15,color='red',fontweight='bold',fontfamily='serif')
plt.ylabel('y axis--->',fontsize=15,color='red',fontweight='bold',fontfamily='serif')

plt.plot(x,y,label='x vs y',color='magenta',marker='^',linestyle='-.',linewidth=4 , ms=10)
plt.plot(x,z,label='x vs z',color='blue',marker='o',linestyle=':',linewidth=4 , ms=10)

plt.legend()
plt.show()
```

```python
x = np.arange(1,11)
y = np.arange(5,15)
plt.gca().set_facecolor('black')

plt.title('1st line graph',fontsize=20,color='green',fontweight='bold',fontfamily='serif')
plt.xlabel('x axis--->',fontsize=15,color='red',fontweight='bold',fontfamily='serif')
plt.ylabel('y axis--->',fontsize=15,color='red',fontweight='bold',fontfamily='serif')

plt.plot(x,y,'o:b',label='x vs y')
# line style: --- , : , -- , -.
# colors: r,g,b,w,k,m,y,c
plt.legend()
plt.show()
```

```python
x = np.arange(1,11)
y = np.arange(5,15)
z = np.arange(21,41,2)

plt.subplot(1,2,1)
plt.plot(x,y,label='x vs y',color='magenta',marker='^',linestyle='-.',linewidth=4 , ms=10)

plt.legend()

plt.subplot(1,2,2)
plt.plot(x,z,label='x vs z',color='blue',marker='o',linestyle=':',linewidth=4 , ms=10)

plt.legend()
plt.suptitle('subplots',fontsize=20,color='red',fontweight='bold')
plt.show()
```

```python
student = ['a' , 'b' , 'c']
marks = [10,23,21]
plt.title('bar graph')
plt.xlabel('student')
plt.ylabel('marks')
plt.bar(student,marks)
plt.show()
```

```python
student = ['a' , 'b' , 'c']
marks = [10,23,21]
plt.title('bar graph')
plt.xlabel('student')
plt.ylabel('marks')
plt.bar(student,marks,color=['red','green','blue'])
plt.show()
```

```python
student = ['a' , 'b' , 'c']
marks = [10,23,21]
plt.title('bar graph')
plt.xlabel('student')
plt.ylabel('marks')
plt.bar(student,marks,color=['red','green','blue'])
plt.grid(True)
plt.show()
```

```python
student = ['a' , 'b' , 'c']
marks = [10,23,21]
plt.title('bar graph')
plt.ylabel('student')
plt.xlabel('marks')
plt.barh(student,marks,color=['red','green','blue'])
plt.grid(True)
plt.show()
```

```python
# scatter plot
x = np.arange(1,11)
y = np.arange(5,15)
plt.scatter(x,y,color='blue')
plt.show()
```

```python
# histogram
data = [1,3,3,3,3,9,9,5,4,4,8,6,7]
plt.hist(data,align='left',bins=4)
plt.show()
```

```python
# pie chart
import matplotlib.pyplot as plt 
import numpy as np

y = [35,25,25,15]
l = ['apple' , 'banana' , 'cherry' , 'grapes']
plt.pie(y,labels=l,explode=[0.2,0,0,0],autopct='%0.1f%%',colors=['red','pink','blue','orange'])
plt.show()
```
